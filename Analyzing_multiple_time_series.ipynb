```python
# Loading packeges 
import pandas as pd 
import matplotlib.pyplot as plt
from statsmodels.graphics import tsaplots
import statsmodels.api as sm
import seaborn as sns
```

```python
#Loading the data
```

```python
#Read csv file
df= pd.read_csv(
	'data.csv',
    parse_dates=['datestamp'], 
    index_col='datestamp',
)
df.head()
```

```python
# Visualizing the data
```

```python
#Plot settings
%config InlineBackend.figure_format='retina'
plt.rcParams['figure.figsize'] = (18,10)
plt.style.use ('ggplot')

#Ploting all time series in the df
ax= df.plot(
	colormap='Spectral',
    fontsize=10,
    linewidth=0.8
)

# Setting labels and legend
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Unemployment Rate', fontsize=12)
ax.set_title('Unemployment rate of U.S workers by industry', fontsize=15)
ax.legend(
	loc='center left',
    bbox_to_anchor=(1.0,0.5), 
)

#Annoteting plots with vertical lines
ax.axvline(
	'2001-07-01', #Position of the vertical line  
    color='red',
    linestyle='--',
    linewidth=2,
)
ax.axvline('2008-09-01', 
           color='red',
          linestyle='--',
           linewidth=2
          )
plt.show()
```

```python
#Inspecting the data
```

```python
df.describe()
```

```python
# Generate a boxplot viz
ax= df.boxplot(fontsize=10, vert=False)
ax.set_xlabel('Unemployment Percentage')
ax.set_title('Distribution of Unemployment by industry')
plt.show()
```

```python
#Analizing subsets of the data
```

```python
# Perform autocorrelation to measures how the lagged version of the value of a variable is related to the original version of it in a time series

fig= tsaplots.plot_acf(
	df['Agriculture'], lags=24
)
plt.show()
```

```python
# Partial autocorrelation plot of the time series

fig= tsaplots.plot_pacf(
	df['Agriculture'], lags=24
)

plt.show()
```

```python
#Group data by time periods to uncover patterns related with the frequency
```

```python
# Extracting time period of interest 
index_year = df.index.year

# Mean 
df_by_year = df.groupby(index_year).mean()

#Plot the for each period 
ax= df_by_year.plot(fontsize=10, linewidth=1)

# Setting labels and legend 
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Mean unemployment rate', fontsize=12)
ax.axvline(
	2008,
    color='red',
    linestyle='--',
    linewidth=2,
)
ax.legend(
	loc='center left',
    bbox_to_anchor=(1.0,0.5)
)
plt.show()
```

```python
#Decomposing time series into seasonality, trend and noise
```

```python
# Decomposing time series
df_names= df.columns 
df_decomp= {ts: sm.tsa.seasonal_decompose(df[ts]) for ts in df.columns}

# Capturing the seasonal, trend and noise components
seasonal_dict = {ts: df_decomp[ts].seasonal for ts in df_names}
trend_dict= {ts: df_decomp[ts].trend for ts in df_names}
noise_dict= {ts: df_decomp[ts].resid for ts in df_names}

# Creating a df for the dictionaries
seasonality_df= pd.DataFrame.from_dict(seasonal_dict)
trend_df= pd.DataFrame.from_dict(trend_dict)
noise_df = pd.DataFrame.from_dict(noise_dict)

#Removing the label for the index
seasonality_df.index.name = None
trend_df.index.name= None
noise_df.index.name= None 
```

```python
#Ploting individual seasonality, trend or noise 
noise_df['Agriculture'].plot()
```

```python
# Creating a faceted plot of the seasonality_df DataFrame
trend_df[['Agriculture', 'Manufacturing']].plot(
	subplots=True,
    layout=(2,1),
    sharey=False,
    legend=True,
    fontsize=10,
    linewidth=2,
)
plt.suptitle('Seasonality in Agriculture and Manufacturing', size=15)
plt.show()
```

```python
# Visualizing the correlations with a clustermap.Clustermap's are use it to order the data by similarity
```

```python
# Get the correlation matrix
seasonality_corr = seasonality_df.corr(
	method='spearman')

# Customizing clustermap
fig= sns.clustermap(
	seasonality_corr,
	annot=True,
	annot_kws={'size':10},
	linewidth=0.4,
	figsize=(15,10),
)

plt.setp(
	fig.ax_heatmap.xaxis.get_majorticklabels(),
    rotation=90,
)
plt.show()
```
